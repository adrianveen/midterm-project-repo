```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
```

```python
sns.set_theme(style="darkgrid", context='talk')
np.set_printoptions(threshold=np.inf)
```

```python
#import worled energy dataset
wec_df = pd.read_csv('World Energy Consumption.csv')
```

```python
wec_df.head()
```

```python
#years from 1965 to 2021
wec_df = wec_df.loc[(wec_df['year'] >= 1965) & (wec_df['year'] <= 2021), :]
```

```python
wec_df.info()
```

```python
pd.set_option('display.max_columns', None)
wec_df.describe()
```

```python
#remove any rows without iso_code value
wec_df = wec_df.loc[wec_df['iso_code'].notna(), :]
```

```python
wec_df.info()
```

```python
wec_df.reset_index(drop=True, inplace=True)
```

```python
wec_df.head()
```

```python
#columns to remove
to_remove = ['solar', 'wind', 'coal', 'renewables', 'oil', 'gas', 'nuclear', 'hydro', 'biofuel', 'other_', 'low_carbon', 'fossil']
columns_to_remove = [col for col in wec_df.columns if any(word in col for word in to_remove)]
```

```python
wec_df = wec_df.drop(columns_to_remove, axis=1)
```

```python
wec_df.columns
```

```python
wec_df.head()
```

```python
wec_df.describe()
```

```python
#drop rows where population is null
wec_df = wec_df.loc[wec_df['population'].notna(), :]
```

```python
#show rows where GDP is null
wec_df.loc[wec_df['gdp'].isna(), :].sort_values(by='year')['country'].unique()
```

```python
wec_df.info()
```

```python
columns_for_iqr = ['population', 'gdp', 'carbon_intensity_elec', 'electricity_demand', 'electricity_generation', 'electricity_share_energy', 'energy_cons_change_pct', 'energy_cons_change_twh', 'energy_per_capita','energy_per_gdp', 'net_elec_imports','net_elec_imports_share_demand', 'per_capita_electricity','primary_energy_consumption']
```

```python
iqr_values = {}
for col in columns_for_iqr:
    q3, q1 = np.percentile(wec_df[col].dropna(), [75, 25])
    iqr = q3 - q1
    iqr_values[col] = iqr
```
