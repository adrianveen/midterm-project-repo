```python
#function to remove rows with zero's out of the list of supplied columns
def remove_zeros(df, cols):
    df = df.copy()
    for col in cols:
        df = df.loc[df[col] != 0, :]
    return df
```

```python
bad_zero_colns = ['carbon_intensity_elec', 'electricity_demand', 'electricity_generation',
       'electricity_share_energy',
       'energy_per_capita', 
       'energy_per_gdp',
       'greenhouse_gas_emissions',
       'per_capita_electricity']

wec_df_cleaned = remove_zeros(wec_df, bad_zero_colns)
```

```python
#remove countries where the count of GDP values is zero
def remove_null_gdp(df):

    countries_with_zero = [index for index, value in df.groupby('country')['gdp'].count().sort_values().items() if value == 0]
    df = df.loc[~df['country'].isin(countries_with_zero), :]
    
    return df
```
